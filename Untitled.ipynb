```python
from pylab import *
x = np.linspace(-np.pi, np.pi, 256)
print(x)
```

```python
%matplotlib inline
cosx = np.cos(x)
sinx = np.sin(x)
plot(x, cosx)
plot(x, sinx)
show()
```

```python


from pylab import *
x = np.linspace(-np.pi, np.pi, 256)
cosx = np.cos(x)
sinx = np.sin(x)
figure(figsize=(10,6), dpi=80)
plot(x, cosx, color="blue", linewidth=2.5, linestyle="-")
plot(x, sinx, color="red",  linewidth=2.5, linestyle="-")

show()

```
